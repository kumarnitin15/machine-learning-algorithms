## DBSCAN tries to cluster datasets that are not clustered properly by KMeans

```python
from sklearn.datasets import make_moons
from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
```

## Data Preparation

```python
X,Y = make_moons(n_samples=200,noise=0.10)
```

```python
plt.scatter(X[:,0],X[:,1],c=Y)
plt.show()
```

## Applying KMeans

```python
kmeans = KMeans(n_clusters=2)
```

```python
kmeans.fit(X)
```

```python
Ypred = kmeans.labels_
centers = kmeans.cluster_centers_
```

```python
plt.scatter(X[:,0],X[:,1],c=Ypred)
plt.scatter(centers[:,0],centers[:,1],c='red')
plt.title("KMeans is not able to find the patterns appropriately")
plt.show()
```

## Applying DBSCAN

```python
dbs = DBSCAN(eps=0.25)
```

```python
Ypred = dbs.fit_predict(X)
```

```python
plt.scatter(X[:,0],X[:,1],c=Ypred)
plt.title("DBSCAN does much better in finding the right clusters")
plt.show()
```
